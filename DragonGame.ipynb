```python
import random
import time

print("What is your name? ")
name=input()
print("Hello!"+name+". Let's play a game.")

def gameintro():
  print("""You are in a land full of dragons. In front of you there are two cave.
In one cave, the dragon is friendly and will share a tresure with you.
In another cave, the dragon is hungry and angry and it will eat you instatnty......""")


def ChooseCave():
  cave=""
  while cave!="1" and cave!="2":
    print()
    print ("Which cave are you chossing to go into?(1 or 2 )")
    cave=input()
  return cave
print()
def CheckingCave(ChosenCave):
  print()
  print ("You are approaching to the cave...... ")
  time.sleep(2)
  print()
  print("""The very walls of this cave hold memories of those who succumbed to its grasp.
Feel the icy breath of their regrets brush against your skin as you advance,
leaving your own mark among their chilling tales. It is dark and spooky......""")
  print()
  time.sleep(6)
  print("""A monstrous dragon suddenly materializes before your very eyes. He
open his jaws and...... """)
  print()
  time.sleep(3)
  FriendlyDragon= random.randint(1,2)

  if ChosenCave==str(FriendlyDragon):
    print("Hurray!You have got the treasure.")
  else:
    print("Consumed you in a solitary monstrous bite!")
ContinuePlaying="Yes"
print()

while ContinuePlaying=="Yes" or ContinuePlaying=="YES" or ContinuePlaying=="yes":
  gameintro()
  CaveNumber= ChooseCave()
  CheckingCave(CaveNumber)
  print()
  print("Do you want to play again? (Yes or No)")
  ContinuePlaying= input()
  print()

print()

print("Bye Bye!"+ name+". See you again.")
```
